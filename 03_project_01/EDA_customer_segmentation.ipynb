## Subsetting rows

```python
# Visualización de las filas de "Annual_Income" con valores menores a 25000. Organizadas por la misma columan de forma ascendente

dataset[dataset['Annual_Income']<25000].sort_values("Annual_Income", ascending=True)
```

```python
dataset[(dataset["Age"]<20) & (dataset["Gender"] == "Male")]
```

```python
home = dataset["Preferred_Product"] == "Home"
visits = dataset["Monthly_Visits"] > 5
fem = dataset["Gender"] == "Female"

dataset[(home)&(visits)&(fem)].sort_values(["Monthly_Visits", "Annual_Income"])
```

```python
dataset["Preferred_Product"].value_counts()
```

```python
product = ["Groceries", "Home"]
female = ["Female"]
score = dataset["Spending_Score"] > 50


dataset[(dataset["Preferred_Product"].isin(product)) & (dataset["Gender"].isin(female)) & score].sort_values("Preferred_Product")
```
