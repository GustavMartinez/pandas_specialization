## Sorting rows

```python
# Sorting by column "Fare" and showing first 5 rows

dataset.sort_values("Fare").head(5)
```

```python
# Sorting by column "Fare", in descending order, and showing first 5 rows

dataset.sort_values("Fare", ascending=False).head()
```

```python
dataset.sort_values(["Fare", "Age"], ascending=[False, True]).head(10)
```

## Subsetting columns

```python
# Select a vector (column) of the dataset, and show its 5 first values

dataset["Name"].head(5)
```

```python
# Select more than one vector (column) on a dataset, and show its 5 first values

dataset[["Name", "Sex", "Age"]].head(10)
```

## Subsetting rows

```python
# subsetting with logical and comparison operators

dataset[dataset["Age"] > 70]
```

```python
dataset[dataset["Age"] < 10].sort_values("Age")
```

```python
less_than_10 = (dataset["Fare"] > 7) & (dataset["Fare"] < 8)
dataset[less_than_10]
```

```python
dataset[dataset["Embarked"] == "Q"]
```

```python
dataset[(dataset["Age"]>=18) & (dataset["Age"]<=26) & (dataset["Sex"] == "female") & (dataset["Survived"] == True) & (dataset["Embarked"] == "C")].sort_values("Fare")
```
